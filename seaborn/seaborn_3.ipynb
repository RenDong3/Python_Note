```python
#导入模块
#允许图片inline显示
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#Scipy的stats模块包含了多种概率分布的随机变量
from scipy import stats,integrate
```

```python
##############单变量 单特征分析     选用直方图分析 ###################
#生成高斯分布数据  默认为标准高斯分布
x  = np.random.normal(size = 200)
#绘制直方图  不进行核密度估计
sns.distplot(x, kde=False)
```

```python
#指定bins为30  即平均分为30等份
sns.distplot(x, bins = 30, kde=False)
```

```python
# x = np.random.gamma(6, size = 100)
#调用scipy库中的stats模块进行统计拟合
sns.distplot(x, bins = 20, fit=stats.gamma,kde=False)
```

```python
#不设置kde = False会出现两条线
#一条为核密度估计曲线
#一条为拟合曲线
sns.distplot(x, bins = 30, fit = stats.gamma)
```

```python
#根据均值和协方差生成数据
#设定均值mean和协方差 cov
mean, cov = [1, 0], [(1, 0.5), (0.5,1)]
#生成正态分布矩阵数据 共200个
data = np.random.multivariate_normal(mean, cov, size = 200)
#转换为pandas的DataFrame结构  columns 指定列名x,y
df = pd.DataFrame(data= data, columns=['x', 'y'])
df
```

```python
##观测两个变量之间的分布关系最好用散点图
sns.jointplot(x='x', y = 'y',data = df)
```

```python
#生成数据  数据量较大  设定kind可以清晰表示数据的集中区域
x,y = np.random.multivariate_normal(mean, cov, size = 1000).T
#设定整体背景风格
with sns.axes_style('white'):
    #指定kind = 'hex',颜色的深浅用来表示数据的聚集  颜色深的区域  数据比较集中密集
    #设定背景为白色  数据为黑色 对比更明显
    sns.jointplot(x= x, y=y, kind = 'hex',color = 'black')
```

```python
#读入sns库自带的鸢尾花iris数据集
iris = sns.load_dataset('iris')
#pairplot绘制成对特征之间的对应关系  iris共4个特征
#sepal_width sepal_length  petal_width petal_length
#对角线为直方图  表示特征自身的关系  
##其余散点图表示特征两两之间的关系 

#显示单特征的直方图以及特征与特征之间的散点图
sns.pairplot(iris)
```
