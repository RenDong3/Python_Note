```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
##############回归分析绘图
#导入seaborn自带数据  tips
#记录顾客消费账单 以及小费情况
tips = sns.load_dataset('tips')
tips.head()
```

```python
#regplot()和lmplot()都可以绘制回归图  推荐使用regplot()
#得到账单和小费之间的关系
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.regplot( x = 'size', y = 'tip',data = tips)
```

```python
#对size数据进行轻微抖动 位移  使其符合回归的数据特性
#x_jitter设置偏移量
sns.regplot(x = 'size', y = 'tip',data = tips, x_jitter=0.1)
```
