```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl
%matplotlib inline
```

```python
                                 ##################多变量分析绘图 ##############
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
tips.head()
```

```python
#默认jitter = True
#根据类别画图  比如一周七天 进行类别绘图
sns.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
#指定hue属性为sex
# sns.swarmplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', hue = 'time', data = tips)
```

```python
## 盒图 ##
# ** IQR即统计学概念四分位距  第1/4位和第3/4位之间的距离
# N= 1.5IQR  如果一个值大于Q3+N 或者小于Q1+N, 就是比3/4位加上N还有大 ,比1/4位减去N还要小
# 称之为离群点
```

```python
sns.boxplot(x = 'day', y = 'total_bill',hue = 'time', data = tips)
###最下面横线为1/4位  最上面横线为3/4位   中间为2/4位  最上面或者最下面的菱形表示离群点 
```

```python
#小提琴图
sns.violinplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
```

```python
#指定split属性  左边蓝色代表male  右边代表female
sns.violinplot(x= 'day', y = 'total_bill', hue = 'sex', split = True, data = tips)
```

```python
#组合小提琴图和树图
#color设置颜色 避免两种图都是用默认颜色 
#alpha设置透明度
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y= 'total_bill', data = tips, alpha = 0.5,color = 'k')
```

```python
#表示数据的集中显示趋势选择条形图
```

```python
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic)
#描述不同船舱等级   男性和女性的获救存活率
```

```python
#点图可以更好的描述变化差异
```

```python
#对于不同船舱等级   女性存活率都要比男性要高  
sns.pointplot(x = 'sex', y = 'survived', hue = 'class', data = titanic)
help(sns.pointplot)
```

```python
#添加属性  使图像更美观
#markers指定数据的符号标识  linestyles指定两点之间连线的格式  如使用虚线还是实线等
sns.pointplot(x = 'class', y = 'survived', hue= 'sex',data = titanic, markers=["o", "x"],linestyles=["-", "--"])
```

```python
#多功能绘图函数
#不指定属性kind  默认为折线图
sns.factorplot(x= 'day', y = 'total_bill', hue = 'smoker', data = tips)
```

```python
#指定kind属性为bar  绘制条形图
sns.factorplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips, kind  = 'bar')
```

```python
#指定属性为swarm 绘制树图
#额外添加col  指定列的属性
sns.factorplot(x = 'day', y = 'total_bill' ,hue = 'smoker', data = tips, col = 'time', kind = 'swarm')
```

```python
sns.factorplot(x = 'time', y = 'total_bill', hue = 'smoker' ,data = tips,
               col = 'day', size = 6, aspect = 0.4, kind = 'box')
#size 指定大小  kind指定为盒图  aspect设置长宽比
```
