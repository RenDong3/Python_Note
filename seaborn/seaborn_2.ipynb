```python
#导入模块
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize' : (6,6)})
```

```python
                                  ##########调色板 ##########
##   color_palette()能传入任何matplotlib支持的颜色
##   color_palette()不写参数则传入默认颜色
##   set_color_palette()设置所有图的颜色
```

```python
###离散型 色彩板 
#        **分类画板**
```

```python
current_pale = sns.color_palette()
sns.palplot(current_pale)
```

```python
#         **圆形画板**
# 当你有六个以上的分类要区分时,最简单的方法是在一个圆形空间中画出均匀间隔的颜色(这样的色调会保持亮度和饱和度不变),这是大多数的当他们需要
# 使用比默认六种颜色更多时采用的默认方案.
# 最常用功的方法是使用hls的颜色空间,这是GRB的一个简单转换
```

```python
#传入10种分类颜色
sns.palplot(sns.color_palette('hls', 10))
```

```python
data = np.random.normal(size = (20, 8))
sns.boxplot(data = data, palette= sns.color_palette('hls', 8))
```

```python
#hls_plaette()函数来控制颜色的亮度和饱和
#l - lightness 控制亮度
#s - saturation 控制饱和度
sns.palplot(sns.hls_palette(8, l = .8,s= .9))
```

```python
# sns.palplot(sns.color_palette('hls', 8))
#Paired 设置颜色成对出现  深蓝--浅蓝  深黄--浅黄....
sns.palplot(sns.color_palette('Paired'))
```

```python
######### 使用xkcd颜色来命名颜色　＃＃＃＃＃＃＃＃＃＃＃#
##　　使用ｘkｃｄ使用固定颜色
#xkcd包含了一套对随机ＲＧＢ颜色的命名，产生了９５４个可以随时通过ｘｔｃd＿ｒｇｂ字典调用颜色
```

```python
#需要指定颜色名称选取特定颜色
#可以去官网查找  各颜色名称  lw代表linewidth即线宽
plt.plot([0, 1], [0, 3], sns.xkcd_rgb['pale red'], lw  = 3)
plt.plot([0,1], [0,2], sns.xkcd_rgb['medium green'], lw = 2)
plt.plot([0,1], [0,1], sns.xkcd_rgb['denim blue'], lw = 4)
```

```python
######连续型色彩板
#色彩随数据变化   比如数据越来越重要  颜色越来越深
```

```python
#'Blues'  颜色默认由浅到深
sns.palplot(sns.color_palette('Blues'))
```

```python
###颜色名称后添加_r后缀  则反转颜色顺序 即由深到浅
sns.palplot(sns.color_palette('Reds_r'))
```

```python
##cubehelix_palette()调色板
#色调线性变换
```

```python
#method 1
sns.palplot(sns.color_palette('cubehelix', 8))
```

```python
#method 2
sns.palplot(sns.cubehelix_palette(8, start=7, rot=0.5))
```

```python
##light_palette() 和dark_palette()调用定制连续调色板
#绿色从浅到深
sns.palplot(sns.light_palette('green'))
```

```python
#红色从深到浅
sns.palplot(sns.dark_palette('red'))
```

```python
#reverse = True则颜色变为从浅到深
sns.palplot(sns.dark_palette('red', reverse=True))
```

```python
##指定颜色空间即  RGB色彩空间
sns.palplot(sns.light_palette((200, 80, 70), input = 'husl'))
```
