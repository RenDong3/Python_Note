```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
tips.head()
```

```python
############FacetGrid使用方法 ##############
#通过FaceGrid()绘制数据集中子图   指定col =time ,time分为dinner和lanch两个属性
#所以绘制出两个空白图
sns.FacetGrid(tips, col = 'time')
```

```python
g = sns.FacetGrid(tips, col = 'time')
#map()函数内指定hist 绘制条形图  tip表示小费情况
g.map(plt.hist, 'tip')
```

```python
#sex  有male 和female两个图  hue属性为smoker
g = sns.FacetGrid(tips, col = 'sex', hue = 'smoker')
#根据tip 和total_bill作为x,y绘制散点图 
g.map(plt.scatter, 'tip', 'total_bill', alpha = 0.8)
#添加图例 即hue属性
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col = 'sex', row = 'time')
g.map(sns.regplot, 'tip', 'total_bill', fit_reg = True,color = 'k')
```

```python
#对FacetGrid 图像大小和整体纵横比进行设置
g = sns.FacetGrid(tips, col = 'day', size = 4, aspect = 0.8)
g.map(sns.barplot, 'sex', 'total_bill')
```
