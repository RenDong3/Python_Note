```python
#导入模块
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
                                     ######### 整体布局风格设置 ############
#定义函数绘制sin函数
def plotsin(filp = 1):
    #横坐标为[1,14]之间的100个点
    x = np.linspace(1, 14, 100)
    #绘制6条sin函数   幅值不同
    for i in range(1, 7):
        plt.plot(x, np.sin(x)*(7-i)*filp)
```

```python
#set()不传入任何参数
#使用matplotlib默认风格
sns.set()
plotsin()
```

```python
###5种主题风格
##-----darkgrid
# -----whitegrid
# -----dark 
# -----white
# -----ticks
```

```python
#背景风格为白色横线
sns.set_style('whitegrid')
#正态分布 默认为标准正态分布 即mean = 0, std= 1
data = np.random.normal(size = (20,5))
sns.boxplot(data = data)
```

```python
#背景风格为纯白色
sns.set_style('white')
sns.boxplot(data = data)
```

```python
#横纵坐标带有毛刺
sns.set_style('ticks')
plotsin()
```

```python
                                  ############## 风格细节设置 ###########
plotsin()
#去除上方及右方的线
sns.despine()
#隐藏左侧轴线
sns.despine(left = True)
```

```python
#offset为图像离轴线的距离
plotsin()
sns.despine(offset=50)
```

```python
                      ############不同子图设置不同背景分风格
#with 指定一种整体风格
#with内部执行都为datkgrid
#with外部执行为默认风格
with sns.axes_style("darkgrid"):
    plt.subplot(2,1,1)
    plotsin()
    
plt.subplot(2, 1, 2)

plotsin(-2)
```

```python
sns.set()
```

```python
#设置上下文风格   及图形尺寸大小
#context风格有'paper'  --'talk'--- 'notebook' 三种
sns.set_context('talk')
plt.figure(figsize= (8,6))
plotsin()
```

```python
#设置字体 及线宽
sns.set_context(context='paper', font_scale= 2, rc = {'lines.linewidth' : 4})
plotsin()
```
