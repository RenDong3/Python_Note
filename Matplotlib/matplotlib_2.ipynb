```python
#导入模块
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#数据读取
unrate = pd.read_csv('UNRATE.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
                    ########## 绘制子图 ###########
#设置figure()对象
fig = plt.figure()
#设置axes对象  子图设置
##1 -->2
##-->3 -->4
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,4)
plt.show()
```

```python
#设置figure尺寸4x4
fig = plt.figure(figsize=(4,4))
#添加子图
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
#绘制子图  每一个子图显示不同的图像
ax1.plot(np.random.random(5), np.arange(5))
ax2.plot(np.arange(4)*2, np.arange(4))
plt.show()
```

```python
data1 = unrate[0:12]
data2 = unrate[12:24]

fig = plt.figure(figsize = (6,3))
plt.plot(data1['DATE'], data1['VALUE'],c = 'red')
plt.plot(data2['DATE'], data2['VALUE'], c = 'blue')

plt.show()
```

```python
#调用python datatime模块
#提取月份作为x轴标签
unrate['month'] = unrate['DATE'].dt.month

data1 = unrate[0:12]
data2 = unrate[12:24]

plt.plot(data1['month'], data1['VALUE'], c = 'red')
plt.plot(data2['month'], data2['VALUE'], c= 'blue')

plt.show()
```

```python
#提取月份作为x轴标签
unrate['month'] = unrate['DATE'].dt.month

data1 = unrate[0:12]
data2 = unrate[12:24]

plt.plot(data1['month'], data1['VALUE'], c = 'red',label = '1948')
plt.plot(data2['month'], data2['VALUE'], c= 'blue',label = '1949')
#添加图例
plt.legend(loc = 'best')
# help(plt.legend)
plt.show()
```

```python
#完整图像显示
fig = plt.figure(figsize=(6,4))

unrate['month'] = unrate['DATE'].dt.month
data1 = unrate[0:12]
data2 = unrate[12:24]

plt.plot(data1['month'], data1['VALUE'], c = 'red', label = '1948')
plt.plot(data2['month'], data2['VALUE'], c = 'blue', label = '1949')

plt.xticks(rotation = 45)
plt.xlabel('Months')
plt.ylabel('Unemployer rate')
plt.title('Monthly unemployment trends, 1948-1949')
plt.legend(loc = 'best')

plt.show()
```
