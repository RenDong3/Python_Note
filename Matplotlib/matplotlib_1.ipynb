```python
#导入模块
import pandas as pd
import matplotlib.pyplot as plt
```

```python
                         ####  绘制折线图  ######
#读取数据
unrate = pd.read_csv('UNRATE.csv')
#转为标准时间形式  1948/01/01 ---> 1948-01-01
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
#显示1948年数据
unrate.head(12)
```

```python
plt.plot()
plt.show()
```

```python
#切片操作提取前12行数据  即1948年全部数据
data = unrate[0:12]
#以DATE年份作为横坐标  VALUE作为纵坐标
plt.plot(data['DATE'], data['VALUE'])
plt.show()
```

```python
plt.plot(data['DATE'], data['VALUE'])
#横坐标刻度标签 角度变换45度
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.plot(data['DATE'],data['VALUE'])
plt.xticks(rotation = 70)
#横轴标签
plt.xlabel('Months')
#纵轴标签
plt.ylabel('Unemployer rate')
#标题
plt.title('Monthly unemployment trends, 1948')
plt.show()
```
