```python
#按照属性后缀分类

#列属性名称存为list
col_names = food_info.columns.tolist()
print col_names

#存储以(g)作为后缀结尾的属性名称
g_name = []

for c in col_names:
    if c.endswith("(g)"):
        g_name.append(c)
        
g_name = food_info[g_name]
print '----------------------------------------------------------------------------------------------------我是可爱的分割线'
print g_name.head(3)
```
